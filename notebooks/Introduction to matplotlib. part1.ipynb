## Matplotlib example worflow

```python
# step0 import matplotlib and get ready fro plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[10,20,30,40]

# 2. setup plot
fig,ax=plt.subplots(figsize=(10,10)) # width, height

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title=" a simple plot",
      xlabel='x-axis',
      ylabel='y-axis')

# 5. save and share
fig.savefig('sample-plot.png')
```

## Making figures with NumPy arrays

We want :
- Line plot
- Scatter plot
- Bar plot
- Histogram
- Subplots

```python
import numpy as np
```

```python
# create data
x=np.linspace(0,10,100)
x[:10]
```

```python
# plot the data and create a line plot

fig,ax=plt.subplots()
ax.plot(x,x**2)

# the deafult .plot() is a line plot
```

```python
# use same data to make a scatter plot

fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# another scatter plot

fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# make a plot from dictionary
nut_butter_prices={'Almond butter':10,
                  'Peanut butter':8,
                  'Cashew butter':12}

fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title='Nut butter prices',
      ylabel='Price ($)');
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

## Histograms amd subplots

```python
# make some data
x=np.random.randn(1000)

fig,ax=plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# option 1
# tuple way
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))

# plotting to each axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

The pyplot API is generally less-flexible than the object oriented API and in general, we try to use the object-oriented interface over pyplot interface and aslo using `plt.subplots()`, we can make a single plot or multiple subplots. Therefore flexible.

```python
# subplots option 2
# list indx way

fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x*3)
ax[0,1].scatter(np.random.random(100),np.random.random(100))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));          
             
```
