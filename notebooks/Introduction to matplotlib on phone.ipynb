```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_sales=pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv')
car_sales
```

```python
# remove $ sign from price column and change data type to int

car_sales['Price']=car_sales['Price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
car_sales['Price']=car_sales.Price.astype(int)
```

```python
car_sales['Price']
```

```python
# adding a sales date column to dataframe
car_sales['Sales Date']=pd.date_range('1/02/2024',periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales
```

```python
# plot the total sales over sales date

# we plot using the dataframe
car_sales.plot(y='Total Sales',x='Sales Date')
```

```python
# scatter plot
car_sales.plot(x='Odometer (KM)',y='Price',kind='scatter')
```

```python
# bar plot of no of doors vs make of car

car_sales.plot(y='Doors',x='Make',kind='bar')
```

## Plotting from df5

### Which one to use? (pyplot vs OO method)

- When plotting something quickly, such as a simple plot, it's okay t0 use pyplot.
- For more advanced plots, use the OO method

```python
heart_disease.head()
```

```python
# perform data analysis of a subset of df whose age is > 50

# we use boolean indexing
over_50=heart_disease[heart_disease['age']>50]
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
# a scatter plot of age vs cholestrol with color of target columh

# pyplot method
# plotting directly from the dataframe
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO me6hod mixed with pyplot method

fig,ax=plt.subplots(figsize=(10,6))

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
# ax.set_xlim([45,100]);
```

When plotting in OO method from pandas df, we include an ax parameter whem we plot.

And this method is not completely OO.

```python
# OO method from scratch

fig,ax=plt.subplots(figsize=(10,6))

# plot 
scatter=ax.scatter(x=over_50['age'],
                  y=over_50['chol'],
                  c=over_50['target'])

# customise the plot
ax.set(title='Heart disease and Cholestrol levels',
      xlabel='Age',
      ylabel='Cholestrol')

# create legend
ax.legend(*scatter.legend_elements(),title='Target')

# add a horizontal line
# sezrch how to add a horizontal line refer doc and try it out
# the line rrpresents the mean vakue of cholestrol
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

*scatter.legend_elements() means grab the values from scatter and unpack the legend elements which is gonna query the c parameter from the scatter plot and cr4ate legend based on the elements taregt colujn has in teh dataframe.

 Scatter plot of heart diseasd snd trestbps

```python
over_50.head()
```

```python
# create fig
fig,ax=plt.subplots(figsize=(10,6))

# plot hearst disease vs trestbps
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['trestbps'],
                   c=over_50['target']
                  )

# customjse the plot
ax.set(title='Heart disease and trest levels',
      xlabel='Age',
      ylabel='Trest bps')
# create a legend
ax.legend(*scatter.legend_elements(),title='Target');
```

```python
# subplots

fig,ax=plt.subplots(nrows=2,
                   ncols=1,
                   figsize=(10,6))

# plot 
scatter=ax[0].scatter(x=over_50['age'],
                  y=over_50['chol'],
                  c=over_50['target'])

# customise the plot
ax[0].set(title='Heart disease and Cholestrol levels',
      xlabel='Age',
      ylabel='Cholestrol')

# create legend
ax[0].legend(*scatter.legend_elements(),title='Target')

# add a horizontal line
# sezrch how to add a horizontal line refer doc and try it out
# the line rrpresents the mean vakue of cholestrol
ax[0].axhline(over_50['chol'].mean(),
          linestyle='--');

#plot 2
# plot hearst disease vs trestbps
scatter2=ax[1].scatter(x=over_50['age'],
                   y=over_50['trestbps'],
                   c=over_50['target']
                  )

# customjse the plot
ax[1].set(title='Heart disease and trest levels',
      xlabel='Age',
      ylabel='Trest bps')
# create a legend
ax[1].legend(*scatter2.legend_elements(),title='Target');

# hline 
ax[1].axhline(over_50.trestbps.mean(),
              ls='--');
```

```python
# subplots of age, chol and thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
# customise
ax0.set(title='Heart disease and Cholestrol levels',
        xlabel='Age',
        ylabel='Cholestrol')

# add a legend
ax0.legend(*scatter.legend_elements(),title='Target')

# hline
ax0.axhline(over_50['chol'].mean(),
            linestyle='--');

# add data to ax1
# plot 2 age vs thalach
scatter1=ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])

# customise 
ax1.set(title='Heart disease and Max heart rate',
        xlabel='Age',
        ylabel='Max heart rate')

#legend
ax1.legend(*scatter1.legend_elements(),title='Target')

#hline
ax1.axhline(over_50['thalach'].mean(),
           ls='--');

# add a title to figure
fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold');
```

```python
#set the style for plot
plt.style.use('seaborn-v0_8-whitegrid')
fig,ax=plt.subplots(figsize=(10,6))

# plot hearst disease vs trestbps
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['trestbps'],
                   c=over_50['target'],
                   cmap='winter' # this changes the colour scheme 
                  )

# customjse the plot
ax.set(title='Heart disease and trest levels',
      xlabel='Age',
      ylabel='Trest bps')
# create a legend
ax.legend(*scatter.legend_elements(),title='Target');
```

```python
# subplots of age, chol and thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter')
# customise
ax0.set(title='Heart disease and Cholestrol levels',
        xlabel='Age',
        ylabel='Cholestrol')
# change x lims
ax0.set_xlim([50,80])
# add a legend
ax0.legend(*scatter.legend_elements(),title='Target')

# hline
ax0.axhline(over_50['chol'].mean(),
            linestyle='--');

# add data to ax1
# plot 2 age vs thalach
scatter1=ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                    cmap='winter')

# customise 
ax1.set(title='Heart disease and Max heart rate',
        xlabel='Age',
        ylabel='Max heart rate')
# change ax1 x axis lims
ax1.set_xlim([50,80])
ax1.set_ylim([80,200])
#legend
ax1.legend(*scatter1.legend_elements(),title='Target')

#hline
ax1.axhline(over_50['thalach'].mean(),
           ls='--');

# add a title to figure
fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold');
```

We use x lim and y lim to clean the borders and give a clean single line or border.
